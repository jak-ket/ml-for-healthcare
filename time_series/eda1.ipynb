# EDA 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load the time series data 
train_data = pd.read_csv("data/ptbdb_train.csv", header=None)
test_data = pd.read_csv("data/ptbdb_test.csv", header=None)
```

```python
train_data.shape
```

```python
y_train = train_data[187].values
y_test = test_data[187].values
x_train = train_data.iloc[:, :-1].values
x_test = test_data.iloc[:, :-1].values
```

### Plot example time series

```python
# plot the first 5 rows 
plt.figure(figsize=(14, 6))
plt.plot(x_train[0], label= y_train[0])
plt.plot(x_train[1], label= y_train[1])
plt.plot(x_train[2], label= y_train[2])
plt.plot(x_train[3], label= y_train[3])
plt.plot(x_train[4], label= y_train[4])
plt.legend()
plt.title("ECG Heartbeat")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.show()
```

### Missing values

```python
# check for missing values and inf values
print("Missing values in training data: ", train_data.isnull().sum().sum())
print("Missing values in testing data: ", test_data.isnull().sum().sum())
print("Infinite values in training data: ", train_data.isin([float('inf'), float('-inf')]).sum().sum())
print("Infinite values in testing data: ", test_data.isin([float('inf'), float('-inf')]).sum().sum())
```

### Examine distribution of labels

```python
# Examine distribution of labels
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

# Plot distribution of labels (Training)
sns.histplot(y_train, ax=axes[0], bins=2)
axes[0].set_title("Distribution of Labels (Training) n = {}".format(len(y_train)))

# Plot distribution of labels (Testing)
sns.histplot(y_test, ax=axes[1], bins=2)
axes[1].set_title("Distribution of Labels (Testing) n = {}".format(len(y_test)))

plt.tight_layout()
plt.show()
```

```python
# look at quantitaive distribution of labels
print("Training Data persentage of positive labels: ", sum(y_train) / len(y_train))
print("Testing Data persentage of positive labels: ", sum(y_test) / len(y_test))
```
